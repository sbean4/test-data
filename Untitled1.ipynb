```python
Wage=pd.read_csv("Data/Wage.csv")
```

```python
Wage_group_one=Wage[['year','wage']]
```

```python
Wage_group_one=Wage_group_one.groupby(['year'],as_index= False).mean()
Wage_group_one
```

```python
Wage_group_one.plot(kind='scatter', x='year', y='wage', figsize=(10, 6), color='darkblue')

plt.title('Wages over time')
plt.xlabel('Year')
plt.ylabel('Wage')

plt.show()
```

```python
x = Wage_group_one.year      # year on x-axis
y = Wage_group_one.wage     # total on y-axis
fit = np.polyfit(x, y, deg=1)

fit
```

```python
Wage_group_one.plot(kind='scatter', x='year', y='wage', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x,fit[0]*x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2008, 110))

plt.show()

# Print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```
